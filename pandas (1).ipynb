```python
# import pandas
import pandas as pd
```

```python
# Create pandas series using list

data = [10, 20, 30, 40]
series = pd.Series(data)
print(series)
```

```python
# Create series using a custom index
data = [1,2,3,4]
i = ['A', 'B', 'C', 'D']
series = pd.Series(data, index=i)
print(series)
```

```python
## Create a pandas series using dictionary
data = {'a': 10, 'b': 20, 'c': 30}
series = pd.Series(data)
print(series)
```

```python
series.replace(20,40)
```

```python
# create series using numpy array
import numpy as np

data = np.array([100, 200, 300])
series = pd.Series(data, index=['a','b','c'])
print(series)
```

```python
# Create  a dataframe with the array
import pandas as pd
data = {"weight":[66,75,84,96,48,71],\
        "height":[156,165,186,165,174,167]}
bmi = pd.DataFrame(data)
bmi
```

```python
# Create a  new BMI column
bmi["BMI"] = bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
# Insert new record in the 
import numpy as np
bmi.loc[3] = [78,np.nan,np.nan]
bmi
```

```python
# Insert 
bmi.loc[5] = [np.nan, 65, np.nan]
bmi
```

```python
# Check Column-wise  sum of NaN  values(missing values)
bmi.isnull().sum()
```

```python
# WEIGHT
bmi["weight"] = bmi["weight"].fillna(65)
bmi
```

```python
#HEIGHT
bmi["height"] = bmi["height"].fillna(75)
bmi
```

```python
# update BMI_NEW
bmi["BMI_New"] = bmi["weight"]/(bmi["height"]/100)**2
bmi
```

```python
# Drop BMI column
bmi.drop("BMI", axis = 1)
```

```python
# drop  BMI column permanently
bmi.drop("BMI", axis = 1, inplace=True)
bmi
```
